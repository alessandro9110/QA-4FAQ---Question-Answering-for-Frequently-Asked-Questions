```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook
import pandas as pd
import csv
```

```python
#epoch;acc;loss;val_acc;val_loss
df = pd.read_csv("log.csv", sep = ";", encoding = 'utf8', quotechar = '"',error_bad_lines=False)

df.head()
```

```python
# summarize history for accuracy
plt.plot(df['acc'])
plt.plot(df['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.savefig('plot_acc.png')
plt.show()


  
# summarize history for loss
plt.plot(df['loss'])
plt.plot(df['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.savefig('plot_loss.png')
plt.show()


# summarize history for loss
plt.plot(df['mean_squared'])
plt.plot(df['val_mean_squared'])
plt.title('model mean_squared')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.savefig('plot_mean_squared.png')
plt.show()
```
